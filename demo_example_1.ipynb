```python
import nest_asyncio
import asyncio
from telethon import TelegramClient
import os

# –ü–∞—Ç—á–∏–Ω–≥ –∞–∫—Ç–∏–≤–Ω–æ–≥–æ —Ü–∏–∫–ª–∞ —Å–æ–±—ã—Ç–∏–π
nest_asyncio.apply()
# –£—Å—Ç–∞–Ω–æ–≤–∏ —Å–≤–æ–∏ API ID –∏ Hash, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å –Ω–∞ https://my.telegram.org/
api_id = '23538457'
api_hash = 'e951f8edb6d75602a5772d31312f936e'

# –£–∫–∞–∑—ã–≤–∞–µ–º –ø—É—Ç—å –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è —Ñ–∞–π–ª–∞ —Å–µ—Å—Å–∏–∏
session_file = os.path.join('sessions', 'my_session')
client = TelegramClient(session_file, api_id, api_hash)

async def fetch_messages(channel_username, limit=10):
    # –ü—Ä–æ–≤–µ—Ä—è–µ–º –ø–æ–¥–∫–ª—é—á–µ–Ω–∏–µ
    if not client.is_connected():
        print("–ü–æ–¥–∫–ª—é—á–∞–µ–º—Å—è –∫ Telegram...")
        await client.connect()

    # –ü—Ä–æ–≤–µ—Ä—è–µ–º –∞–≤—Ç–æ—Ä–∏–∑–∞—Ü–∏—é
    if not await client.is_user_authorized():
        print("–ê–≤—Ç–æ—Ä–∏–∑–∞—Ü–∏—è –Ω–µ–æ–±—Ö–æ–¥–∏–º–∞. –í—ã–ø–æ–ª–Ω—è–µ–º –≤—Ö–æ–¥...")
        await client.start()

    # –ü—ã—Ç–∞–µ–º—Å—è –ø–æ–ª—É—á–∏—Ç—å —Å–æ–æ–±—â–µ–Ω–∏—è
    try:
        messages = []
        async for message in client.iter_messages(channel_username, limit=limit):
            if message.text:
                messages.append({
                    'id': message.id,
                    'date': message.date.strftime('%Y-%m-%d %H:%M:%S'),
                    'text': message.text
                })
        return messages
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ –ø–∞—Ä—Å–∏–Ω–≥–µ –∫–∞–Ω–∞–ª–∞ {channel_username}: {e}")
        return []

async def save_messages_to_file(messages, filename="parsed_messages.txt"):
    # –û—Ç–∫—Ä—ã–≤–∞–µ–º —Ñ–∞–π–ª –¥–ª—è –∑–∞–ø–∏—Å–∏ –≤ —Ä–µ–∂–∏–º–µ 'a' (append)
    with open(filename, 'a', encoding='utf-8') as file:
        for msg in messages:
            file.write(f"[{msg['date']}] –°–æ–æ–±—â–µ–Ω–∏–µ ID {msg['id']}:\n")
            file.write(f"{msg['text']}\n")
            file.write('-' * 50 + '\n')

async def main():
    channel_username = "studygrants"  # –£–∫–∞–∂–∏ username –∫–∞–Ω–∞–ª–∞
    print(f"–ü–∞—Ä—Å–∏–Ω–≥ –ø–æ—Å–ª–µ–¥–Ω–∏—Ö {10} —Å–æ–æ–±—â–µ–Ω–∏–π –∏–∑ –∫–∞–Ω–∞–ª–∞ {channel_username}...")
    messages = await fetch_messages(channel_username, limit=10)

    # –í—ã–≤–æ–¥–∏–º —Å–æ–æ–±—â–µ–Ω–∏—è
    for msg in messages:
        print(f"[{msg['date']}] –°–æ–æ–±—â–µ–Ω–∏–µ ID {msg['id']}: {msg['text']}")
        print('-' * 50)
    # –ó–∞–ø–∏—Å—ã–≤–∞–µ–º —Å–æ–æ–±—â–µ–Ω–∏—è –≤ —Ç–µ–∫—Å—Ç–æ–≤—ã–π —Ñ–∞–π–ª
    if messages:
        await save_messages_to_file(messages)
        print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ —Ñ–∞–π–ª parsed_messages.txt")
    else:
        print("–ù–µ—Ç —Å–æ–æ–±—â–µ–Ω–∏–π –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è.")

    # –ó–∞–∫—Ä—ã–≤–∞–µ–º –∫–ª–∏–µ–Ω—Ç–∞
    await client.disconnect()

    # # –£–¥–∞–ª—è–µ–º —Ñ–∞–π–ª —Å–µ—Å—Å–∏–∏
    # if os.path.exists(f"{session_file}.session"):
    #     print("–û—á–∏—â–∞–µ–º —Ñ–∞–π–ª —Å–µ—Å—Å–∏–∏...")
    #     os.remove(f"{session_file}.session")
    # else:
    #     print("–§–∞–π–ª —Å–µ—Å—Å–∏–∏ –Ω–µ –Ω–∞–π–¥–µ–Ω.")

if __name__ == "__main__":
    import asyncio
    asyncio.run(main())
```
